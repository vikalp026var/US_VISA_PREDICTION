```python
numeric_feature=[feature for feature in df.columns if df[feature].dtype!='O']
categorical_feature=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
print("We have {} numeric features :{}".format(len(numeric_feature),numeric_feature))
```

```python
print("We have {} categorical features :{}".format(len(categorical_feature),categorical_feature))
```

```python
for col in categorical_feature:
    print(df[col].value_counts(normalize=True)*100)
```

```python
#BiVariate Anlaysis 
#Multivariate ANlaysis 
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Univariate Analaysis of Numerical Features",fontsize=20,fontweight='bold',alpha=0.8)

for i in range(0,len(numeric_feature)):
    plt.subplot(2,2,i+1)
    sns.kdeplot(x=df[numeric_feature[i]],color='blue')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

```python
categorical_feature
```

```python
# categorical_feature.remove('case_id')
plt.figure(figsize=(15,10))
plt.suptitle("Univariate Analaysis of Categorical Features",fontsize=20,fontweight='bold',alpha=0.8)

for i in range(0,len(categorical_feature)):
    plt.subplot(3,3,i+1)
    sns.countplot(x=df[categorical_feature[i]])
    plt.xlabel(categorical_feature[i])
    plt.tight_layout()
```

```python
##Multivariate 
discrete_features=[feature for feature in numeric_feature if len(df[feature].unique())<=25]
continuous_features=[feature for feature in numeric_feature if len(df[feature].unique()) > 25]
print('We have {} discrete features : {}'.format(len(discrete_features), discrete_features))
print('\nWe have {} continuous_features : {}'.format(len(continuous_features), continuous_features))
```

```python
#Chi square Test 
```

```python
from scipy.stats import chi2_contingency
chi2_test=[]
for feature in categorical_feature:
    if chi2_contingency(pd.crosstab(df['case_status'],df[feature]))[1]<0.05:
        chi2_test.append("Reject the Null Hypothese")
    else:
        chi2_test.append("Fail to Reject the Null Hypothesis")
result=pd.DataFrame(data=[categorical_feature,chi2_test]).T
result.columns=['Column','Hypothesis Result']
result
```
